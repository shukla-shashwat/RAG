```python
## create a simple txt file
import os   
os.makedirs("../data/text_file", exist_ok=True)
```

```python
sample_text={
   "../data/text_file/python.txt": """Python Basics

Python is a high level, interpreted programming language known for its clear syntax
and readability. It supports multiple programming paradigms such as procedural,
object oriented, and functional programming. Python is widely used for rapid
development, scripting, and building production grade applications.

Key Features of Python:
1. Simple and readable syntax that is easy to learn.
2. Large standard library for handling files, networking, data processing, and more.
3. Cross platform support, runs on Windows, macOS, and Linux.
4. Huge ecosystem of third party libraries for web development, data science,
   machine learning, automation, and more.
5. Interactive shell for quick testing and experimentation.

Common uses include web development, data analysis, machine learning, automation,
scripting, API development, and scientific computing.""",
    
    
    "../data/text_file/machine_learning.txt": """Machine Learning Basics

Machine learning is a branch of artificial intelligence that allows systems to learn 
and improve from data without being explicitly programmed. It focuses on building 
models that can analyze information, identify patterns, and make decisions or 
predictions based on experience.

Types of Machine Learning:
1. Supervised Learning: Learning using labeled data.
2. Unsupervised Learning: Discovering patterns in unlabeled data.
3. Reinforcement Learning: Learning through rewards and penalties.

Common applications include image recognition, speech processing, fraud detection, 
medical diagnosis, and recommendation systems.
"""
}

for file_path, content in sample_text.items():
    with open(file_path, "w", encoding="utf-8") as f:
        f.write(content)    

print("Sample text files created.")       
```

```python
### TextLoader
# from langchain.document_loaders import TextLoader
from langchain_community.document_loaders import TextLoader 
loader = TextLoader("../data/text_file/python.txt", encoding="utf-8")
data = loader.load()
print(data)
```

```python
### DirectoryLoader
from langchain_community.document_loaders import DirectoryLoader
loader = DirectoryLoader("../data/text_file", 
                         glob="*.txt",
                         loader_cls=TextLoader,
                         loader_kwargs={"encoding": "utf-8"}, 
                         show_progress=True)
data = loader.load()
print(data)
```

```python
### pdfDirectoryLoader
from langchain_community.document_loaders import PyPDFLoader,PyMuPDFLoader
loader = DirectoryLoader("../data/pdf", 
                         glob="*.pdf",
                         loader_cls=PyMuPDFLoader,
                         show_progress=True)
pdf_document = loader.load()
print(pdf_document)
pdf_document
```
